```python
def get_artist_ids(track_ids):
    artist_ids = set()
    for track_id in track_ids:
        result = sp.track(track_id)['artists']
        for artist in result:
            artist_ids.add(artist['id'])
    
    return list(artist_ids)


def get_related_artists(artist_id):
    related_artists = []
    result = sp.artist_related_artists(artist_id)['artists']
    for artist in result:
        related_artists.append(artist['id'])
    artists = set()
    # print(related_artists)
    try:
        artists.update(related_artists[:2])
        related_artists.sort(reverse = True, key = get_artist_popularity)
        artists.update(related_artists[:2])
        artists.update(random.sample(related_artists[3:], k = 2))
    except:
        return []
        
    return list(artists)


def get_top_tracks(artist_id):
    top_tracks = []
    result = sp.artist_top_tracks(artist_id)['tracks']
    for track in result:
        top_tracks.append(track['id'])
    tracks = set()
    try:
        tracks.update(random.sample(top_tracks[:5], k = 2))
        tracks.update(random.sample(top_tracks[6:], k = 3))
    except:
        tracks.update(top_tracks)
    
    return list(tracks)


def get_artist_popularity(artist_id):
    return sp.artist(artist_id)['popularity']



```

```python
def get_recommended_tracks(track_ids):
    recommended_tracks = set()
    artist_ids = get_artist_ids(track_ids)
    for artist_id in artist_ids:
        related_artists = get_related_artists(artist_id)
        
        for artist in related_artists:
            top_tracks = get_top_tracks(artist)
            recommended_tracks.update(top_tracks)
            
    return random.sample(list(recommended_tracks), k = 100)

recommended_tracks = get_recommended_tracks(track_ids)
recommended_tracks         
        
```
