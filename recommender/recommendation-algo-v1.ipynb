```python
artist_ids = []
for track_id in track_ids:
    result = sp.track(track_id)
    for artist in result['artists']:
        artist_ids.append(artist['id'])
        
artist_ids
```

```python
def get_popularity(artist):
    return artist['popularity']

related_artist_ids = set()
for artist_id in artist_ids:
    try:
        result = sp.artist_related_artists(artist_id)['artists']
        related_artist_ids.add(result[0]['id'])
        related_artist_ids.add(result[1]['id'])
        result[2:].sort(reverse = True, key = get_popularity)
        related_artist_ids.add(result[2]['id'])
        related_artist_ids.add(result[3]['id'])
        for related_artist in random.sample(result[4:], k = 2):
            related_artist_ids.add(related_artist['id'])
    except:
        pass
        
related_artist_ids.update(artist_ids)
```

```python
for related_artist_id in related_artist_ids:
    result = sp.artist(related_artist_id)
    print(result['name'])
```

```python
recommended_track_ids = set()
for related_artist_id in related_artist_ids:
    result = sp.artist_top_tracks(related_artist_id)['tracks']
    for track in result:
        recommended_track_ids.add(track['id'])
print(len(recommended_track_ids))
recommended_track_ids = list(recommended_track_ids)
recommended_track_ids
```
